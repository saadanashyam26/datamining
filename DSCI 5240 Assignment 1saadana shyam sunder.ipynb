5. Explore the relationship between variable 'KM' and 'Price' using statistics and a graph. Describe your findings.

```python
toyato_cars['KM'].value_counts()
toyato_cars['Price'].value_counts()
```

```python
km_statistics = toyato_cars['KM'].describe()
price_statistics = toyato_cars['Price'].describe()
km_statistics
price_statistics
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(toyato_cars['KM'], toyato_cars['Price'], alpha=0.5, color='green')
sn.regplot(x='KM',y='Price',data=toyato_cars,line_kws={'color': 'red'})
plt.title('Relationship between KM and Price for Toyota Corollas')
plt.xlabel('KM (Accumulated Kilometers)')
plt.ylabel('Price (Euros)')
plt.grid(True)

```

```python
correlation_Km_and_price = toyato_cars['KM'].corr(toyato_cars['Price'])
correlation_Km_and_price 
```

Interpretation:

Scatter Plot: The scatter plot displays the relationship between 'KM' and 'Price'. Each point represents a car, with the x-coordinate indicating accumulated kilometers ('KM') and the y-coordinate indicating the sale price ('Price'). This visualization helps identify any trends or patterns.
Descriptive Statistics: The descriptive statistics for 'KM' and 'Price' provide insights into the central tendency and variability of each variable.

Analyzing the scatter plot:
 there is a negative correlation(-0.5699601645337193), you may observe a downward trend in the scatter plot, suggesting that as the accumulated kilometers increase, the price tends to decrease.
 
The parameter alpha controls the transparency of the data points. The value of alpha ranges from 0 to 1, where 0 is fully transparent (invisible), and 1 is fully opaque (not transparent). Setting alpha=0.5 means that the data points in the scatter plot will be semi-transparent.

interpretation:

Similar Mean: The fact that the mean prices are the same could indicate that while there is a difference in the starting points of the price distributions, the central tendency (mean) is not affected by this difference.
